```python
y=x.loc[:,['A','B']].values
y
```

```python
wcss=[]
for j in range(1,11):
    kmeans=KMeans(n_clusters=j,init='k-means++')
    kmeans.fit(y)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2,color='blue',marker='8')
plt.xlabel('K Value')
plt.ylabel('WCSS')
plt.show()
```
